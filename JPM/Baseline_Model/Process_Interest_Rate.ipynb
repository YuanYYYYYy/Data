```python
import pandas as pd
import numpy as np
```

```python
import os
path = "/content/gdrive/MyDrive/JPM/Interest Rate/"
os.chdir(path)
os.listdir(path)
```

```python
files_US=os.listdir(path+"/US")
files_UK=os.listdir(path+'/UK')
files_US
```

```python
def process_index(data):
  year=data.split('年')[0]
  month=data.split('年')[1].split('月')[0].zfill(2)
  date=data.split('年')[1].split('月')[1].split('日')[0].zfill(2)
  return str(year+month+date)
```

```python
Year=['Rate_30_','Rate_10_','Rate_5_','Rate_2_',"Rate_6m","Rate_3m"]
```

```python
dicter=[]
for Y in Year:
  lister=[]
  for f in files_US:
    if Y in f:
      print(Y)
      tmp=pd.read_csv(path+"US/"+f)
     # print(tmp)
      tmp.Date=tmp.Date.apply(process_index)
      lister.append(tmp)
  data=pd.concat(lister,axis=0)
  #data.Date=data.Date.apply(process_index)
  data.sort_values(by='Date',inplace=True)
  data.index=data.Date
  data[Y.split('_')[1]]=data["Close"]
  dicter.append(data[[Y.split('_')[1]]])
```

```python
data.tail(60)
```

```python
pd.concat(dicter,axis=1).dropna().to_csv(path+'US/Combined.csv')
```
