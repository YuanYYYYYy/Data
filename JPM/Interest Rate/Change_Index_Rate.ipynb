```python
import pandas as pd

for i in files_list[:2]:
  print(i)
  
  data=pd.read_csv(path+i)
  data.Date=data.Date.apply(get_Date)
  data.to_csv(path+i)
```

```python
def get_Date(data):
  year=data.split('年')[0]
  #print(year)
  month=data.split('年')[1].split('月')[0].zfill(2)
  #print(data.split('年')[1].split('月')[0].zfill(2))
  Date=data.split('年')[1].split('月')[1].split('日')[0].zfill(2)
  #print(data.split('年')[1].split('月')[1].split('日')[0].zfill(2))
  return year+month+Date
```
