```python
import os
from tqdm import tqdm
import pdfplumber
fail_list=[]
month={'Jan':'01','Feb':'02','Mar':'03','Apr':'04','May':'05','Jun':'06','Jul':"07",'Aug':'08','Sep':"09",'Oct':'10','Nov':'11','Dec':'12'}
for file in tqdm(range(len(file_list))):
    try:
      index=file
      print(file)
      file=file_list[file]
      #if file.replace('pdf', 'txt') not in txt_files:
      with pdfplumber.open(path + file) as pdf:
          content = ''
          num = 0
          for page in pdf.pages:
              num += 1
              text = page.extract_text()
              text=text.replace('\n',' ')
              text=text.replace('\xa0','')
              text=text.strip()
              if text:
                  content += text
          m=''
          for i in month.keys():
            if (i) in file:
              m=month[i]
              break
          file_date=file.split('.')[0][-4:]+m+file[:2]
          print(file_date)
          if content:
            with open("/content/gdrive/MyDrive/JPM/UK_data/UK_Speech/"+file_date+'.txt', 'w') as f:
              f.write(content)
    except:
      fail_list.append(file_list[index])
```
